```python
def smooth(y, window=100):
    return np.convolve(y, np.ones(window)/window, mode='valid')

# Example:
# plt.plot(x, y)  # original (noisy) plot
plt.figure(figsize = [15,5])
plt.plot(x[99:], smooth(y, window=100))  # smoothed plot

plt.axvline(1420.5,lw=1, color="black") # y = 0
plt.xticks(np.linspace(min(x), max(x), 15))

plt.savefig("smoothed.png", dpi=300)
plt.show()
```

```python
def interpolate(d1,d2,alpha):
    return (1-alpha)*d1+alpha*d2
```

```python
metadata = dict(title='Progression', artist='LouisRis')
writer = PillowWriter(fps=15,metadata=metadata)

fig = plt.figure(figsize=[15,5])



with writer.saving(fig, 'skyScanInterpolation.gif',100):
    for i in range(29):
        plt.cla()
        plt.plot(data[i,99:,0],smooth(data[i,:,1]- background[:,1],window=100))
        plt.xticks(np.linspace(min(x), max(x), 15))
        plt.ylim(0,7e-5)
        writer.grab_frame()
        
        plt.cla()
        plt.plot(data[i,99:,0],interpolate(smooth(data[i,:,1]- background[:,1],window=100),smooth(data[i,:,1]- background[:,1],window=100),0.5))
        plt.xticks(np.linspace(min(x), max(x), 15))
        plt.ylim(0,7e-5)
        writer.grab_frame()
        
    plt.plot(data[29,99:,0],smooth(data[29,:,1]- background[:,1],window=100))
    plt.xticks(np.linspace(min(x), max(x), 15))
    plt.ylim(0,7e-5)
    writer.grab_frame()
plt.close()
```
